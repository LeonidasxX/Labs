```python
# 3
columns = ['Month_Number', 'Total_Profit']
df = pd.read_csv('./Company.csv', usecols=columns)
plt.plot(df.Month_Number, df.Total_Profit)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()
```

```python
#4
plt.plot(df.Month_Number, df.Total_Profit, linestyle='dotted',color='red', label='Profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right', title='Profit')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.show()
```

```python
plt.plot(df.Month_Number, df.Total_Profit, 'o',linestyle='dotted',color='red', label='Profit', linewidth=3,)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right', title='Profit')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.show()
```

```python
# Extra
columns = ['Month_Number', 'Total_Profit', 'Car','Bus', 'Bike']
df = pd.read_csv('./Company.csv', usecols=columns)
plt.plot(df.Month_Number, df.Total_Profit, label='Profit')
plt.plot(df.Month_Number, df.Car, label='Car')
plt.plot(df.Month_Number, df.Bus, label='Bus')
plt.plot(df.Month_Number, df.Bike, label='Bike')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend()
plt.show()
```
